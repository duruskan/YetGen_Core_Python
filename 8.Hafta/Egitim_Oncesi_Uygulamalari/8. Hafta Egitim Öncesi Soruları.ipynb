# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
# 1.Problemin Çözümünü Buraya Yazınız
def defterdar():
    with open("urunler.txt", "w", encoding="utf-8") as file:
        devamMi="t"
        while devamMi=="t":
            ad= input("Ürün ismini ve fiyatını (ad, fiyat) formatında giriniz:")
            file.write(ad+"\n")
            devamMi = input("Ekleme yapmak istiyorsanız t tuşuna basın?")
defterdar()
file = open("urunler.txt", "r",encoding="utf-8")
print(file.read())
file.close()   
````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python
# 2.Problemin Çözümünü Buraya Yazınız
def degistirici():
    with open("urunler.txt", "r+", encoding="utf-8") as file:
        eskiKelime = input("Değiştirmek istediğiniz ürünün ismini ve fiyatını (ad, fiyat) formatında giriniz:")
        yeniKelime = input("Yeni ürünün ismini ve fiyatını (isim, fiyat) formatında giriniz:")
        plain_text = file.readlines()
        for (indeks,satir) in enumerate(plain_text):
            if eskiKelime in satir:
                plain_text[indeks] = plain_text[indeks].replace(eskiKelime, yeniKelime)
    with open("urunler.txt", "w", encoding="utf-8") as file:
        file.writelines(plain_text)    
degistirici()
file = open("urunler.txt", "r",encoding="utf-8")
print(file.read())
file.close()   
````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
# 3.Problemin Çözümünü Buraya Yazınız
class Vehicle:
    def __init__(self, name, speed, mileage):
        self.name= name
        self.speed= speed
        self.mileage= mileage
class Bus(Vehicle):
    def __init__(self, name, speed, mileage):
        super().__init__(name, speed, mileage)
toPrint = Bus("School Volvo", "180", "12" )
print(f"Name: {toPrint.name} Speed: {toPrint.speed} Mileage: {toPrint.mileage}")
````
