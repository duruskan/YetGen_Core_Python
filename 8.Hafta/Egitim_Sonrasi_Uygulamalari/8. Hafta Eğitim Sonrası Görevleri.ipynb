# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
with open("kullanici.txt", "w", encoding="utf-8") as file:
    nested_kullanici= []
    kisi_sayisi = int(input("Kaç adet giriş yapmak istiyorsunuz?"))
    zayif= []
    normal_kilolu = []
    fazla_kilolu = []
    obez = []
    morbid_obez =[]
    for i in range(1,kisi_sayisi+1):
        isim_soyisim_for = input(f"{i}.kişinin isim ve soyismini giriniz:")
        yas_for =int(input(f"{i}.kişinin yaşını giriniz:"))
        boy_for= float(input(f"{i}.kişinin boyunu giriniz:"))
        kilo_for = float(input(f"{i}.kişinin kilosunu giriniz:"))
        if kilo_for/(boy_for**2) <18.5:
            nested_kullanici.append([isim_soyisim_for, yas_for, boy_for,kilo_for,kilo_for/(boy_for**2),"zayıf"])
            zayif.append(nested_kullanici[i-1])
            temp_list = nested_kullanici[i-1]

            temp_str= f"{temp_list[0]},{temp_list[1]},{temp_list[2]},{temp_list[3]},{temp_list[4]},{temp_list[5]} "
            file.writelines(temp_str+"\n")
        elif 25>kilo_for/(boy_for**2) >=18.5:
            nested_kullanici.append([isim_soyisim_for, yas_for, boy_for,kilo_for,kilo_for/(boy_for**2),"normal kilolu"])
            normal_kilolu.append(nested_kullanici[i-1])
            temp_list = nested_kullanici[i-1]

            temp_str= f"{temp_list[0]},{temp_list[1]},{temp_list[2]},{temp_list[3]},{temp_list[4]},{temp_list[5]} "
            file.writelines(temp_str+"\n")
        elif 30>kilo_for/(boy_for**2) >=25:
            nested_kullanici.append([isim_soyisim_for, yas_for, boy_for,kilo_for,kilo_for/(boy_for**2),"fazla kilolu"])
            fazla_kilolu.append(nested_kullanici[i-1])
            temp_list = nested_kullanici[i-1]

            temp_str= f"{temp_list[0]},{temp_list[1]},{temp_list[2]},{temp_list[3]},{temp_list[4]},{temp_list[5]} "
            file.writelines(temp_str+"\n")
        elif 40>kilo_for/(boy_for**2)>=30:
            nested_kullanici.append([isim_soyisim_for, yas_for, boy_for,kilo_for,kilo_for/(boy_for**2),"obez"])
            obez.append(nested_kullanici[i-1])
            temp_list = nested_kullanici[i-1]

            temp_str= f"{temp_list[0]},{temp_list[1]},{temp_list[2]},{temp_list[3]},{temp_list[4]},{temp_list[5]} "
            file.writelines(temp_str+"\n")
        elif kilo_for/(boy_for**2)>=40:
            nested_kullanici.append([isim_soyisim_for, yas_for, boy_for,kilo_for,kilo_for/(boy_for**2),"ileri derecede obez"])
            morbid_obez.append(nested_kullanici[i-1])
            temp_list = nested_kullanici[i-1]

            temp_str= f"{temp_list[0]},{temp_list[1]},{temp_list[2]},{temp_list[3]},{temp_list[4]},{temp_list[5]} "
            file.writelines(temp_str+"\n")
        
file = open("kullanici.txt", "r",encoding="utf-8")
print(file.read())
file.close()  
print(obez)
print(normal_kilolu)
print(zayif)
print(fazla_kilolu)
print(morbid_obez)
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
class YetBar ():
    def __init__(self, sayi):
        self.sayi=int(sayi)
    def menu_goruntuleme(self):
        print("""Menü:
        
Muz: 4 TL
Çilek:4 TL
Elma: 2 TL
Kivi: 5 TL
Ananas:9 TL
Böğürtlen: 9 TL
Mango: 20 TL
Havuç: 4 TL
Salatalık: 3 TL
Pancar: 5 TL
Yoğurt: 5 TL
Fındık: 7 TL
Ceviz: 9 TL
Yeşil Çay: 4 TL
Su: 2 TL
Tarçın: 1 TL
                    
Yukarıdaki ürünlerden en çok 5 tanesini içeceğinize ekleyebilirsiniz.""")
    def satin_alma(self):
        menu_dict = {"muz": 4, "çilek": 4, "elma":2,"kivi":5,"ananas":9,"böğürtlen":9,"mango":20,"havuç":4,"salatalık:":3,"pancar":5,"yoğurt":5,"fındık":7,"ceviz":9,"yeşil çay":4,"su":2,"tarçın":1}
        secim_list=[]
        i=0
        while i in range(0,self.sayi):
            secim= input(f"İçeceğinize eklemek istediğiniz {i+1}.malzemenin ismini giriniz. İşlemi iptal etmek için q'ya basınız").lower()
            if secim in menu_dict.keys():
                secim_list.append(secim)
                print(f"{i+1}.seçiminiz:{secim}: {menu_dict[secim_list[i]]} TL")
                i=i+1
            elif secim=="q":
                break
            else:
                print("Böyle bir malzeme bulunamadı.")
                i=i
        toplam=0
        for i in range(0,len(secim_list)):
            toplam = toplam + menu_dict[secim_list[i]]
        onay= input(f"""
                        Toplam tutar:{toplam} 
                        Bu işlemi onaylıyorsanız 1'e onaylamıyorsanız 0'a basınız.""")
        if onay=="1":
            from datetime import date
            import time
            tarih = date.today()
            alım_tarih= tarih.strftime("%d/%m/%Y")
            with open("satis_kayit.csv","a",encoding="utf-8") as file:
                file.writelines(f"Seçimler: {secim_list}, Toplam Tutar: {toplam}, Tarih:{alım_tarih}")
            print("Satın alım başarılı. İçeceğiniz hazırlanıyor...")
            time.sleep(10)
            print("İçeceğiniz hazır.")
        elif onay=="0":
            print("Satın alma işlemi iptal edildi.")
        else:
            print("Yanlış numarayı tuşladınız.")
            
        

def YetBar_SmoothieBank():
    while True:
        islem_talep=input("""YetBar Smoothie Bank'a Hoşgeldiniz:
                            Aşağıdaki işlemlerden birini seçmek için lütfen yanındaki sayıyı tuşlayın:

                            1.Menüyü görüntülemek
                            2.İçecek satın almak
                            
                            Çıkış yapmak için q tuşuna basınız.""")
        if islem_talep=="1":
            YetBar(0).menu_goruntuleme()
        if islem_talep=="2":
            try:
                malzeme_sayi=int(input("Kaç adet malzeme eklemek istiyorsunuz?"))
                islem2=YetBar(malzeme_sayi)
                islem2.satin_alma()
            except ValueError:
                print("Lütfen bir sayı girişi yapınız!")
        if islem_talep=="q":
            print("YetBar Smoothie Kapanıyor... İyi günler dileriz. ")
            break
            
YetBar_SmoothieBank()


        
```
