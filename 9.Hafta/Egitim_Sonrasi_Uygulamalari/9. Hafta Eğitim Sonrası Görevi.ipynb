# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import random
import time
class Kumanda():
    def __init__(self, tvAcikMi, tvSes, kanalListe, currentKanal):
        self.tvAcikMi= tvAcikMi
        self.tvSes= tvSes
        self.kanalListe= kanalListe
        self.currentKanal= currentKanal
    def tvAc(self):
        if self.tvAcikMi=="Kapalı":
            print("Televizyon Açılıyor!")
            self.tvAcikMi= "Açık"
        else:
            print("Televizyon zaten açık.")
    def tvKapa(self):
        if self.tvAcikMi=="Açık":
            print("Televizyon Kapanıyor!")
            self.tvAcikMi= "Kapalı"
        else:
            print("Televizyon zaten kapalı.")
    def sesAyarlari(self):
        while True:
            ses_istek = int(input("Sesi arttırmak için 1, azaltmak için 0 tuşuna basınız, sesi değiştirmek istemiyorsanız 9 tuşuna basınız:"))
            if ses_istek ==1:
                if self.tvSes==10:
                    print(f"Mevcut ses düzeyi: {self.tvSes}. Daha fazla arttırılamıyor.")
                else:
                    self.tvSes = self.tvSes +1
                    print(f"Mevcut ses düzeyi: {self.tvSes}.")
            elif ses_istek==0:
                if self.tvSes==0:
                    print(f"Mevcut ses düzeyi: {self.tvSes}. Daha fazla azaltılamıyor.")
                else:
                    self.tvSes= self.tvSes-1
                    print(f"Mevcut ses düzeyi: {self.tvSes}.")
            else:
                print(f"Mevcut ses düzeyi: {self.tvSes} olarak belirlendi.")
                break
    def kanalEkle(self):
        eklenecekKanal= input("Eklemek istediğiniz kanalın adını giriniz:")
        print("Kanal ekleniyor. Lütfen Bekleyiniz.")
        time.sleep(1)
        self.kanalListe.append(eklenecekKanal)
        print("Eklemek istediğiniz kanal, listenize başarıyla eklendi.")
    def kanalSayisi(self):
        kanalSayi= len(self.kanalListe)
        print(f"Mevcut kanal sayısı: {kanalSayi}")
    def rastgeleKanal(self):
        sayi= random.randint(0,len(self.kanalListe)-1)
        self.currentKanal= self.kanalListe[sayi]
        print(f"Rastgele açılan kanal: {self.currentKanal}")
    def televizyonBilgi(self):
        print(f"TELEVİZYON BİLGİLERİ\nTelevizyon Durumu:{self.tvAcikMi}\nSes Düzeyi:{self.tvSes}\nMevcut Kanal Listesi:{self.kanalListe}\nŞu anda açık olan kanal: {self.currentKanal} ")
print("""
Kumanda ile yapabileceğiniz işlemler aşağıdadır.
Aşağıdaki işlemlerden birini yapabilmek için yan tarafındaki sayıyı tuşlayınız.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri
Çıkış için q tuşuna basınız.""")

while True:
    istek= input("Yapmak istediğiniz işlemi tuşlayınız:")
    if istek=="1":
        kumanda.tvAc()
        kumanda.tvAcikMi="Açık"
    elif istek=="2":
        kumanda.tvKapa()
        kumanda.tvAcikMi="Kapalı"
    elif istek=="3":
        kumanda.sesAyarlari()
        kumanda.tvSes= kumanda.tvSes+1
    elif istek=="4":
        kumanda.kanalEkle()

    elif istek=="5":
        kumanda.kanalSayisi()
    elif istek=="6":
        kumanda.rastgeleKanal()
    elif istek=="7":
        kumanda.televizyonBilgi()
    elif istek=="q":
        print("Kumandadan çıkış yapılıyor.")
        break
    else:
        print("Geçersiz değer girildi. Tekrar deneyiniz.")
    

    
```
